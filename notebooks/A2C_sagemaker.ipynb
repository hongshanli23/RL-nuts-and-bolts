# Training an A2C agent on SageMaker

Fine tune the docker file from aws/deep-learning-containers

```python
import sagemaker
from sagemaker.pytorch import PyTorch

sess = sagemaker.Session()
ecr_image = "578276202366.dkr.ecr.us-west-2.amazonaws.com/pt:cpu"
role="arn:aws:iam::578276202366:role/service-role/AmazonSageMaker-ExecutionRole-20180626T134894"

script = "/home/ubuntu/RL/RL-nuts-and-bolts/examples/a2c_cartpole.py"
output_path = "s3://" + sess.default_bucket() + "/a2c/cartpole"

est = PyTorch(
    image_uri=ecr_image,
    entry_point=script,
    role=role,
    instance_type="ml.c5.2xlarge",
    instance_count=1,
    output_path=output_path,
)

est.fit()
```
